```python
cars = pd.read_csv('mtcars_empty_data.csv')
cars.head()
```

```python
cars.shape
```

```python
cars.info()
```

```python
cars.describe()
```

```python
# Finding the missing values
```

```python
cars.isnull().sum()
```

```python
cars.isnull().sum().sum()
```

```python
cars.loc[1:3,['drat']] = '#'
cars
```

```python
cars.isnull().sum()
```

```python
s='@ # %'.split()
for i in s:
    cars.replace(i,np.nan,inplace=True)
```

```python
# replace
cars['drat'].replace('#',np.nan,inplace=True)
```

```python
cars
```

```python
cars.isnull().sum() #isna()
```

```python
# Handle/fill the nan values
#Methods-> fill the data-> any value, ffill, bfill, mean, median, mode(categorical data)
```

```python
# fillna
cars.fillna(value=100)
```

```python
cars.fillna(method='ffill') # fills with previous value(LOCF)
```

```python
cars.fillna(method='bfill')# fills with next value(NOCB)
```

```python
cars.fillna(cars.mean())
```

```python
cars.fillna(cars.median())
```

```python
cars.fillna(cars['model'].mode())
```

```python
#Dropping rows/columns
cars.dropna()
```

```python
cars.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
cars
```

```python
cars.drop(3,axis=0)
```

```python
cars.dropna(how='all') # delete rows if all values are null
```

```python
cars.dropna(how='any',inplace=True) # delete rows if any value is null
```

```python
cars.reset_index(drop=True)
```

```python
#finding the duplicates
cars.duplicated()
```

```python
#Drop Duplicates
cars.drop_duplicates()
```
