## Object Oriented Plotting
4 Steps
* Create a blank figure object
* Add Axes
* Generate plots within the object
* specify layout & plotting parameters

```python
fig = plt.figure()
ax = fig.add_axes([0.8,0.5,1,1]) # L,B,W,H
ax.plot(x,y)
plt.show()
```

```python
# Setting limits and ticks
fig = plt.figure()
ax = fig.add_axes([0,0,1,1]) # L,B,W,H
#ax.set_xlim([1,3])
#ax.set_ylim([4,16])
#ax.set_xticks([1,3,5])
ax.set_yticks([9,25])
ax.plot(x,y)
plt.show()
```

## Gnerating Multiple Plots


```python
fig = plt.figure()
fig , ax = plt.subplots(3,3) # sub graph of 2R & 3C
ax[0,0].plot(x,y)
ax[0,1].bar(x,y)
ax[0,2].scatter(x,y)
ax[1,0].plot(x,y)
ax[1,1].pie(x)
ax[1,2].barh(x,y)
ax[2,0].plot(x,y)
ax[2,1].pie(x)
ax[2,2].barh(x,y)
plt.savefig('Multiple Plots.png',format='png')
plt.show()
```

## Plot Formatting-> bar,pie,line

```python
plt.bar(x,y,width=[0.2,0.1,0.3,0.4,0.3],color=['cyan','red','green','yellow','violet'])
```

```python
#plt.labels = ['A','B','C','D','E']
plt.pie(x,[0,0.4,0,0,0],labels = ['A','B','C','D','E'])
plt.show()
```

```python
#Line Chart
plt.plot(x,y,lw=4,ls='-.',color='yellow',marker='p',mew=35,mfc='green',mec='cyan')
```
