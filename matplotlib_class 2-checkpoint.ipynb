## Object Oriented Plotting
4 Steps
* Create a blank figure object
* Add Axes
* Generate plots within the object
* specify layout & plotting parameters

```python
fig = plt.figure() # Step1
ax = fig.add_axes([0.8,0.5,0.9,1]) # L,B,W,H # step 2
ax.plot(x,y) # step 3
plt.show()
```

## Generating Multiple Plots


```python
fig = plt.figure(figsize=(10,15))
fig , ax = plt.subplots(2,3) # sub graph of 3R & 3C
ax[0,0].plot(x,y)
ax[0,1].bar(x,y)
ax[0,2].scatter(x,y)
ax[1,0].plot(x,y)
ax[1,1].pie(x)
ax[1,2].barh(x,y)
#ax[2,0].plot(x,y)
#ax[2,1].pie(x)
#ax[2,2].barh(x,y)
plt.tight_layout(pad=3)
#plt.savefig('Multiple Plots.png',format='png')
plt.show()
```

## Gnerating Plot within a Plot


```python
import numpy as np
x=np.arange(1,11)
y = x ** 2
print(x,y)
fig2 = plt.figure()

# Creating main axes - values b/w 0 & 1 
axes_main = fig2.add_axes([0.1, 0.1, 0.9, 0.9]) # [L,B,W,H]

# Creating inset axes - values b/w 0 & 1
axes_inset = fig2.add_axes([0.5, 0.4, 0.2, 0.3]) 

# Change the numbers in axes_inset and see the difference
# Main figure
axes_main.plot(x, y, 'b')
axes_main.set_xlabel('X_label_main')
axes_main.set_ylabel('Y_label_main')
axes_main.set_title('Title_main')

# Inset figure
axes_inset.plot(y, x, 'r')
axes_inset.set_xlabel('X_label_inset')
axes_inset.set_ylabel('Y_label_inset')
axes_inset.set_title('Title_inset')

plt.show()
```

## Plot Formatting-> bar,pie,line

```python
x=np.arange(1,6)
y=x**2
```

```python
plt.bar(x,y,width=[0.2,0.1,0.3,0.4,0.3],color=['cyan','red','green','yellow','violet'])
```

```python
#plt.labels = ['A','B','C','D','E']
plt.pie(x,[0,0.1,0,0,0],labels = ['A','B','C','D','E'])
plt.show()
```

```python
#Line Chart
plt.plot(x,y,lw=4,ls='-.',color='yellow',marker='p',mew=35,mfc='green',mec='cyan')
```
